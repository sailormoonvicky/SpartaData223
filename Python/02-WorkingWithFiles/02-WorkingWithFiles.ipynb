```python
import csv
```

```python
def name_and_email():
    with open ('user_details.csv', 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            list_of_name_email = []
            list_of_name_email.append(row['firstName'])
            list_of_name_email.append(row['lastName'])
            list_of_name_email.append(row['email'])
            print(list_of_name_email)

name_and_email()          
```

```python
with open ('user_details.csv', 'r') as file:
        reader = csv.reader(file)
        next(reader) # skip the header
        for row in reader:
            print(row)
            print("Name is {}, surname is {}".format(row[1],row[2]))
```
